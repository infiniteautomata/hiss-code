Overview
-------

Okay, so it's pretty obvious that the Southern Baptist Convention is facing a slide into oblivion, but I think it's interesting to try and get some idea of what that process might look like in a little more detail. One very simple way to do that is to perform a birth-death simulation - that name is historical by the way, we can use the same tools to model things like religious conversion, not only literal births and deaths.

Alright then, how do we go about simulating the changes in SBC affiliation using our method of choice? The short version is that we need to get estimates for the inflow and outflow of people from the group that we're looking at. We then step through a time sequence and update our population values by applying those flows to our model. After each update step, we can take a look at our results, to get a feel for how everything is humming along.

Of course, in order to build the model, we need to know our population fluxes in and out of the relevant groups. Since we're looking at a religious community, I'm going to argue we need to model four main things:

- Births
- Deaths
- Conversion
- Deconversion

For our purposes, deconversion can refer to leaving the faith entirely, or just swapping out to another denomination - the result is the same for the SBC either way. Additionally, we'll break our overall population down into cohorts on the basis of age, both so we can get a more fine-grained look at what's going on, and because it's becoming increasingly clear that religious belief is undergoing a dramatic shift among younger people.

Model Parameters
-------

Before we dive into code, I want to discuss how the parameters for this model were chosen, to give some insight into my assumptions, and point out a couple of ways this simulation could be refined in the future.

### Initialization

First up is the question of the initial population distribution over the cohorts. This is relatively straightforward to determine - we can get attendance numbers from the [2018 SBC annual report](http://www.sbcec.org/bor/2018/2018SBCAnnual.pdf), and the breakdown across generational categories for Southern Baptists has been documented by the Pew Research Center in their [Religious Landscape Study](http://www.pewforum.org/religious-landscape-study/religious-denomination/southern-baptist-convention/). For this model, I'll be using the SBC's worship attendance numbers, since it seems likely that this is a better reflection of the committed members who are contributing to the life of the organization, as opposed to every name listed on the bare membership rolls.

### Population Effects

To model births, I'm using [UN numbers via World Bank](https://data.worldbank.org/indicator/sp.dyn.cbrt.in) on the per capita birthrate in the US, which was 12 per 1000 people in 2016. Since this is an aggregate number, I'm going to use it as a direct estimate based on population in the model, without attempts to specialize it for age cohorts or other demographic data. It's also worth noting that the birthrate in the US has been declining for a while; this shouldn't have much impact for short simulations, but it is a potential source of error if we want to use this model to project 50 or 100 years in the future.

To account for deaths, I'm using the Social Security Administration's [Actuarial Life Tables](https://www.ssa.gov/OACT/STATS/table4c6.html), with the caveat that I've averaged the rates for males and females each year, and further averaged all year values within a given cohort. I've also set the upper bound on ages to be 95 - as people age out, they'll be removed from the totals entirely. Overall, the impact of the death rates is most important for people at the upper end of the age range, and considering how top-heavy the SBC is right now, it may be a significant factor in determining the trend of their decline. That being said, errors in my rough averages are likely to wash out, especially in the face of my hard boundary for nonagenarians in the simulation.

### Religious Effects

The other source of gains and losses in our particular population of interest has to do with religious conversion and deconversion. For this, I'm using data provided by Pew on [religious switching](http://www.pewforum.org/2015/05/12/chapter-2-religious-switching-and-intermarriage/). There are a couple of issues here - for one, in their item for Baptists they don't specify Southern Baptist, and it's possible that tally includes data from other, more mainline Baptist denominations which may be experiencing different trends. Also, in a somewhat puzzling choice, they represent movement in and out of the religion as a percentage of the overall US population, rather than in terms of the membership numbers of the denomination itself - although this can be corrected with a bit of algebra. 

The result I'm using here is a little spitballed owing to all these factors, but it suggests that that the SBC should experience a loss of some 43% in each coming-of-age group, offset to some extent by the equivalent of 23% of that number switching into the denomination. For the purposes of this simulation, I'm going to model this as removal from the youngest cohort (since other data indicates most deconversions happen by the early 20s) and I'm going to spread the gains out evenly between the middle three cohorts, since most people entering the denomination are likely to be independent adults making a decision about what kind of denomination they want to practice in long-term.

### For Pedants

Just to be totally clear for anyone reading who might have some simulation experience - what this version of the code does is purely deterministic, so it's not really modeling any actual birth-death processes in the stochastic sense. Right now, it's more like a simulation of [chemical kinetics](https://en.wikipedia.org/wiki/Chemical_kinetics) using discrete time steps. Sue me, I'm trained as a physical chemist. However, one potential enhancement in future versions (should they ever exist) is to actually incorporate a random element and sample probabilities for each of the processes, rather than just scooping people in and out of the groups. That said, with the SBC's membership of 5 million people at the moment, I don't anticipate finite-size effects will cause too much deviation from the deterministic model, although it might be something to look at for very long simulations, if the population dwindles far enough.

The Code
-------

Finally, we're here! Now that we have a methodology and details of the model hammered out, it's time to start playing with code (in Python, of course, what else do you expect from your Snek Fren)!

We'll start by loading up a couple of things we'll need later:

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

And now we'll define some variables to get started:

```python
# Let's define our time reference
baseYear = 2018

# SBC's Sunday worship figures as of 2018 (Change it if you like!)
initialPop = 5320488

# The fraction of membership in each of several demographic groups
# Here, the age ranges I'm using are: [ 0-17, 18-33, 34-53, 54-73, 74-95 ]
# Again, you can adjust to taste
demographicRatios = np.array([0.07, 0.09, 0.24, 0.39, 0.21])

demographicBinSizes = np.array([17.-0.+1., 33.-18.+1., 53.-34.+1., 73.-54.+1., 95.-74.+1.])
demographicRanges = ["0-17", "18-33", "34-54", "54-73", "74-95"]

demographicPop = np.floor(demographicRatios * initialPop)

# UN estimate for the US aggregate birthrate  
# Feel free to adjust if you have better data
# I'd love to see a breakdown by region, income, etc.
birthRateUS = 12./1000.

# Net conversion/deconversion rate
# Departures are assumed to be entirely concentrated in the transition 
# to the 18-33 demographic, while additions are spread over the middle groups
# Of course, feel free to tweak it if you like
religiousOutFlux = np.array([0.43, 0.0, 0.0, 0.0, 0.0])
religiousInFlux = np.array([0.0, 0.0767, 0.0767, 0.0767, 0.0])

# Death rates from the US Social Security Administration's actuaries
# It wouldn't be a birth-death simulation without the reaper!
deathRates = np.array([0.0005211, 0.0009954, 0.002692, 0.01306, 0.09895])
```

Okay, time to run the meat of our simulation!

```python
# First, let's decide how long to watch for (this one's entirely up to you)
yearSpan = 20
popScale = 1e6

# Bookkeeping for our final output
populationHistory = [ demographicPop.sum()/popScale ]
summaryFrequency = 5;
periodicYearDemos = demographicPop
yearList = [ int(baseYear) ]

# The master loop!
for year in range(yearSpan):
    # Remove deaths in each cohort
    demographicPop -= np.floor(demographicPop * deathRates)
    
    # Promote everyone who aged into a new group
    transfers = np.floor(demographicPop/demographicBinSizes)
    demographicPop -= transfers
    
    promotions = np.append(0, transfers[0:-1])
    demographicPop += promotions
    
    # Account for deconversions and denominational switches
    demographicPop -= np.floor(transfers * religiousOutFlux)
    demographicPop += np.floor(transfers[0] * religiousInFlux)
    
    # Add in everyone who was conceived this year
    demographicPop[0] += np.floor(birthRateUS * demographicPop.sum())
    
    # Track the numbers for plotting
    yearList.append(int(baseYear + year + 1))
    populationHistory.append(demographicPop.sum()/popScale)
    if ((year+1)%summaryFrequency == 0):
        periodicYearDemos = np.vstack((periodicYearDemos, demographicPop))
```

Now we have the numbers we need! Let's plot some of them:

```python
# First a nice lines + points plot of total population over time 
fig, ax = plt.subplots()
ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

plt.plot(yearList, populationHistory, 'bo')
plt.plot(yearList, populationHistory, 'b')

plt.xlabel("Year")
plt.ylabel("SBC Membership (Millions, Projected)")
plt.title("Simulated Total Membership")

plt.show()

bins = np.arange(5)

# Now let's look at the demographic breakdown for each 5-year segment
for index in range(np.size(periodicYearDemos,axis=0)):
    totalPop = periodicYearDemos[index].sum()
    plt.bar(bins, np.ndarray.tolist(periodicYearDemos[index]/totalPop))
    plt.xticks(bins, demographicRanges)
    plt.xlabel("Age Cohort")
    plt.ylabel("Membership Share (%)")
    yearVal = baseYear + (index * summaryFrequency)
    title = "Simulated Demographics for " + str(yearVal)
    plt.title(title)
    plt.show()
```

Parting Thoughts
-------

Simulation achieved! Now it's time to reflect a bit on our results and see if there's anything interesting lurking in our data.

To my eyes, the major conclusions that jump out from the plots above are that this model forecasts that the SBC is going to continue to lose membership in a somewhat non-linear way over the next twenty years, and that over time the proportion of younger members may tick up slightly from current levels.

The final question, of course, is how much confidence we have in our model. What I've shown here is a pretty simple setup, but the downward trend accords with our expectations from other data. I do find it interesting that in longer simulations I've run using this model (but haven't shown here) suggest that the half-life of the denomination is around 50 years given current trends - the behavior on that timescale actually looks quite reminiscent of a radioactive (or any constant-rate) decay process, and coupled with relative demographic stability may provide some evidence their losses amongst the young are the biggest driver in their evolution. While this isn't shocking if you've been following news about the SBC and Christianity in general, it does emphasize that they aren't quite skating up to demographic collapse in the immediate future. In other words, they're probably going to linger for a while, and they're probably going to be bitter about it. 

Of course, there are many objections that could still be raised against these conclusions - maybe the birthrate used here should only apply to people of childbearing age (instead of assuming a straight per-capita rate over the total population), or maybe it's unreasonable to skim off all their losses from the 18-33 crowd. Maybe there's going to be a tipping point as a result of financial or social factors that will dramatically alter the trajectory of the SBC's decline.
